```python
import pandas as pd
import joblib

X = pd.read_csv("../data/processed/expense_features.csv")
model = joblib.load("../models/isolation_forest.pkl")

X["anomaly_flag"] = model.predict(X)

anomalies = X[X["anomaly_flag"] == -1]
anomalies
```

```python
print(f"Total anomalies detected: {len(anomalies)}")
```
