```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

os.makedirs("../data/processed", exist_ok=True)

df = pd.read_csv("../data/raw/expense_claims.csv")

features = df[["claim_amount", "days_claimed"]]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

processed_df = pd.DataFrame(
    scaled_features,
    columns=["claim_scaled", "days_scaled"]
)

processed_df.to_csv("../data/processed/expense_features.csv", index=False)

processed_df.head()
```
