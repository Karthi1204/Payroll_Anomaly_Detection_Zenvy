```python
import pandas as pd
import joblib
import time
import warnings
warnings.filterwarnings("ignore")

model = joblib.load("../models/isolation_forest.pkl")
df = pd.read_csv("../data/processed/expense_features.csv")

for _, row in df.iterrows():
    prediction = model.predict([row])[0]
    if prediction == -1:
        print("Real-time anomaly detected:", row.to_dict())
    time.sleep(0.3)
```
